# **B) Random Forest:**

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.linear_model import LogisticRegression

rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
ada_clf = AdaBoostClassifier(n_estimators=50, learning_rate=1, random_state=42)
gb_clf = GradientBoostingClassifier(n_estimators=100, random_state=42)
lr_clf = LogisticRegression(max_iter=1000, random_state=42)

ensemble_clf = VotingClassifier(
    estimators=[
        ('rf', rf_clf),
        ('ada', ada_clf),
        ('gb', gb_clf),
        ('lr', lr_clf)
    ],
    voting='soft'
)

ensemble_clf.fit(X_train, y_train)
y_pred_ensemble = ensemble_clf.predict(X_test)

print('Ensemble Model accuracy score: {0:0.4f}'.format(accuracy_score(y_test, y_pred_ensemble)))

cm_ensemble = confusion_matrix(y_test, y_pred_ensemble)
print('Confusion matrix\n\n', cm_ensemble)

print('\nTrue Positives(TP) = ', cm_ensemble[1, 1])
print('\nTrue Negatives(TN) = ', cm_ensemble[0, 0])
print('\nFalse Positives(FP) = ', cm_ensemble[0, 1])
print('\nFalse Negatives(FN) = ', cm_ensemble[1, 0])

sensitivity_ensemble = cm_ensemble[1, 1] / (cm_ensemble[1, 1] + cm_ensemble[1, 0])
print('\nSensitivity: ', sensitivity_ensemble)
```
